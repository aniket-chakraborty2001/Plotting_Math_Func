15. **Plotting the graph of y=x^(1/2n)** 
*Here we fisrt define the function f(x,n) which returns the value x^(1/2n). After that we define a second function named as plot_function, that depends on the input variable n. Finally we use the input() function and feed it to int() function such that the user can manually change the integer values on n and successfully get differnt plots at same plot area.* *italicized text*

* The value of x lies between 0 and infinity
* Here n is any natural number
* As n becomes big, the function 1/2n becomes small
* When n is infinity, the function 1/2n tends to 0
* So, in that case x^0 will be 1
* Which is the limit of the function as n becomes infonity

```python
def f(x,n):
  return (x**(1/(2*n)))


def plot_function(n):
  x=np.linspace(0,10,1000)
  y= f(x,n)
  plt.plot(x,y)
  plt.xlabel('Values of x')
  plt.ylabel('values of f(x,n)')
  plt.title('Graph of the function y=x^(1/(2*n))')
  plt.show()

n=int(input('Enter a  Natural Number:'))
plot_function(n)
```

22. **Here i plot a function f(x) which takes value 1 when x belongs to the interval [1,2]. f(x) takes value -1 when x belongs to the interval [-2,-1]. f(x) will give 0 if x lies in any other interval rather than the previously mentioned closed intervals\**.

```python
def f(x):
  if (x>1) and (x<2):
    return 1
  elif (x<-1) and (x>-2):
    return -1
  else:
    return 0

x=np.linspace(-5,5,100)

y=[]
for i in range(len(x)):
  y.append(f(x[i]))

plt.plot(x,y,color='red',linestyle='--',marker='o')
ax=plt.gca()
ax.set_ylim([-3,3])
plt.axhline(y=0,color='black')
plt.axvline(x=0,color='black')
plt.show()
```
